# Schwefel

```python
t0 = time()
sol, log, sig = es_plus(10, schwefel, 2000, 10000, 1, 200, schwefel_init,
                        logging=True, log_interval=10, init_std=50.005, roulette=True, cx=True)
print '\n', time() - t0, '\n\n'
plot_pop(sig)
```

```python
def schwefel2(X):
    res = np.zeros(X.shape[0])
    res[np.where((np.abs(X) > 500).sum(axis=1))] = -10000000000000000
    res[res == 0] = schwefel(X)[res == 0]
    return res
```

```python
t0 = time()
sol, log, sig = es_plus(10, schwefel2, 200, 1000, 10, 200, schwefel_init,
                        logging=True, log_interval=10, init_std=10.005, roulette=True, cx=True)
print '\n', time() - t0, '\n\n'
plot_pop(sig)
```
