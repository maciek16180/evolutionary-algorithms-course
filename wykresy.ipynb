```python
def plot_log(fname, opt):
    plt.figure(figsize=(15,8))
    f = np.load(fname)
    sol = f['arr_0']
    log = f['arr_1']
    plt.plot(xrange(len(log)), log)
    if opt is not None:
        plt.plot(xrange(len(log)), [-opt]*len(log))
    plt.show()
    print sol
```

```python
def plot_pop(log):
    plt.figure(figsize=(15,8))
    plt.plot(xrange(len(log)), log)
    plt.show()
```
